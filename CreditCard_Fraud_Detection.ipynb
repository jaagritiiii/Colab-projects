Importing Dependencies(required libraries)


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  accuracy_score
```

```python
# loading dataset to pd dataframe
credit_card_data= pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
#num of missing val
credit_card_data.isnull().sum()
```

```python
#counting valid/invalid transactions using value_counts function
credit_card_data['Class'].value_counts()
```

As we observe, it is highly unbalanced dataset; so we need to do some data pre-processing

0->valid transaction

1->invalid/fraudulent transaction

```python
#data segregation
valid= credit_card_data[credit_card_data.Class==0]
fraud= credit_card_data[credit_card_data.Class==1]
```

```python
print(valid.shape)
print(fraud.shape)
```

```python
#statistics of transactions
valid.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comapre values for both
credit_card_data.groupby('Class').mean()
```

To deal with the unbalanced data, we will use a method called UNDER-SAMPLING

Build sample dataset containing similar distribution of normal/fraud transactions

Fraudulent transactions-> 162

```python
valid_sample= valid.sample(n=162)
```

concatenating 2 Dataframes
(using axis=0, dataframes will be added one by one)

```python
new_dataset=pd.concat([valid_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting this data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

Logistic Regression model for Binary Classification problems as imported

```python
model= LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
# accuracy on training data
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test data is : ', test_data_accuracy)
```

Since the accuracy score on training and test data is almost equal, the model is neither overfitted nor underfitted. Thus, we have created an ML model to detect Fraudulent Credit Card Transactions
